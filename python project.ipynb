```python
# Stacked Bar Chart for Team vs. Position
pivot_table = df.pivot_table(index='Position', columns='Team', aggfunc='size')

pivot_table.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title("Team and Position Stacked Bar Chart")
plt.ylabel("Number of Employees")
plt.xlabel("Position")
plt.legend(title="Team")
plt.tight_layout()
plt.show()
```

```python
# Define age groups
bins = [20, 30, 40, 50, 60]
labels = ['20-30', '31-40', '41-50', '51-60']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group by age group
age_group_distribution = df['Age Group'].value_counts()

# Display the age group distribution
print(age_group_distribution)
```

```python
age_group_distribution.plot(kind='pie', autopct='%1.1f%%', title="Age Group Distribution")
plt.ylabel('')
plt.show()
```

```python
# Line Chart for Age Group Distribution
plt.figure(figsize=(7, 6))
age_group_distribution.sort_index().plot(kind='line', marker='*', color='green')
plt.title("Age Group Trends")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```
